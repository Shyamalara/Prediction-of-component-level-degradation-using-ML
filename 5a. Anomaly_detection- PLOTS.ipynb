```python
## function for defining variables like ps1_1, ps1_2 for 17 sensors with different sampling rates

def col(n, var):
    l = []
    for i in range(1,n):
        temp = str(var) + '_' + '%d' %i
        l.append(temp)
    return l
```

```python
# Calculate IQR for the 1st principal component (pc1)
q1_ps1, q3_ps1 = X['PS1'].quantile([0.25, 0.75])
iqr_ps1 = q3_ps1 - q1_ps1
```

```python
# Calculate upper and lower bounds for outlier for pc1
lower_ps1 = q1_ps1 - (1.5*iqr_ps1)
upper_ps1 = q3_ps1 + (1.5*iqr_ps1)
```

```python
# Filter out the outliers from the pc1
X['anomaly_pc1'] = ((X['PS1']>upper_ps1) | (X['PS1']<lower_ps1)).astype('int')
```

```python

# Let's plot the outliers from pc1 on top of the sensor_11 and see where they occured in the time series
a = X[X['anomaly_pc1'] == 1] #anomaly
plt.figure(figsize=(18,6))
plt.plot(df['PS1'], color='blue', label='Normal')
plt.plot(df['PS1'], linestyle='none', marker='X', color='red', markersize=0.5, label='Anomaly')
plt.xlabel('Date and Time')
plt.ylabel('Sensor Reading')
plt.title('PS1 Anomalies')
plt.legend(loc='best')
plt.show();
```

```python
def anomaly_detect(sensor):
    # Calculate IQR for the 1st principal component (pc1)
    q1, q3 = X[sensor].quantile([0.25, 0.75])
    iqr = q3 - q1
    
    # Calculate upper and lower bounds for outlier for pc1
    lower = q1 - (1.5*iqr)
    upper = q3 + (1.5*iqr)
    
    X['anomaly_pc1'] = ((X[sensor]>upper) | (X[sensor]<lower)).astype('int')
    
    a = X[X['anomaly_pc1'] == 1] #anomaly
    ax = plt.figure(figsize=(18,6))
    plt.plot(X[sensor], color='blue', label='Normal')
    plt.plot(X[sensor], linestyle='none', marker='X', color='red', markersize=5, label='Anomaly')
    plt.xlabel('Cycle')
    plt.ylabel('Sensor Reading')
    plt.title(sensor + ' Anomalies')
    plt.legend(loc='best')
    plt.show()
    
    #return ax 
    
    
    
```

### Using Z-score

```python
cols = list(X.columns) 

for col in cols:
    col_zscore = col + '_zscore'
    X[col_zscore] = (X[col] - X[col].mean())/X[col].std(ddof=0)
```

```python
X.drop(['anomaly_pc1_zscore'], axis = 1, inplace = True)
```

```python
X.drop(['anomaly_pc1'], axis = 1, inplace = True)
```

```python
# Filter out the outliers from the pc2
X['ps1_outlier'] = ((X['PS1_zscore']> 3) | (X['PS1_zscore']< -3)).astype('int')
```

```python

a = X[X['PS1_zscore'] == 1] #anomaly
plt.figure(figsize=(18,6))
plt.plot(df['PS1'], color='blue', label='Normal')
plt.plot(df['PS1'], linestyle='none', marker='X', color='red', markersize=0.5, label='Anomaly')
plt.xlabel('Date and Time')
plt.ylabel('Sensor Reading')
plt.title('PS1 Anomalies')
plt.legend(loc='best')
plt.show();
```

```python
def outlierzscore(sensor, zscore):
    # Filter out the outliers from the pc2
    X[zscore] = ((X[zscore]> 3) | (X[zscore]< -3)).astype('int')
    
    a = X[X['PS1_zscore'] == 1] #anomaly
    plt.figure(figsize=(18,6))
    plt.plot(X[sensor], color='blue', label='Normal')
    plt.plot(X[sensor], linestyle='none', marker='X', color='red', markersize=0.5, label='Anomaly')
    plt.xlabel('Date and Time')
    plt.ylabel('Sensor Reading')
    plt.title( sensor +' Anomalies')
    plt.legend(loc='best')
    plt.show();
    
```
